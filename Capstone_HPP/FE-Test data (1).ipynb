```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
train_df=pd.read_csv('train.csv')
train_df.head()
```

```python
test_df=pd.read_csv('test.csv')
test_df.head()
```

Handle categorical missing values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(test_df[feature].isnull().mean(),4)))
```

```python
def replace_cat_feature_test(test_df, features_nan):
    data = test_df.copy()
    data[features_nan] = data[features_nan].fillna('Missing_Test')
    return data

test_df = replace_cat_feature_test(test_df, features_nan)
test_df.isnull().sum()
test_df
```

Handle numerical missing values


```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(test_df[feature].isnull().mean(),4)))
```

```python
for feature in numerical_with_nan:
    median_value = test_df[feature].median()  # use train median
    # create missing indicator
    test_df[feature+'nan'] = np.where(test_df[feature].isnull(), 1, 0)
    test_df[feature].fillna(median_value, inplace=True)
    test_df[numerical_with_nan].isnull().sum()
```

```python
test_df[numerical_with_nan].isnull().sum()
```

```python
test_df.head(5)
```

Transform temporal features

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    test_df[feature]=test_df['YrSold']-test_df[feature]
```

```python
test_df.head()
```

```python
test_df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
# list of numerical variables
numerical_features = [feature for feature in test_df.columns if test_df[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
test_df[numerical_features].head()
```

```python
for feature in numerical_features[:-1]:  # exclude 'SalePrice' since test doesn't have it
    test_df[feature] = np.log(test_df[feature])
```

```python
test_df.head()
```

Handle rare categorical levels

```python
categorical_features=[feature for feature in test_df.columns if test_df[feature].dtypes=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp = train_df.groupby(feature)['SalePrice'].count()/len(train_df)
    temp_df = temp[temp>0.01].index
    test_df[feature] = np.where(test_df[feature].isin(temp_df), test_df[feature], 'Rare_var')
```

Label encoding based on train set mapping

```python
for feature in categorical_features:
    labels_ordered = train_df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    test_df[feature] = test_df[feature].map(labels_ordered)
    # if some new category is in test but not in train, fill with -1
    test_df[feature] = test_df[feature].fillna(-1)
```

#Feature Scaling

```python
test_df.replace([np.inf, -np.inf], np.nan).isnull().sum()
```

```python
for feature in numerical_features:
    test_df[feature] = np.log1p(test_df[feature])
```

```python
test_df.replace([np.inf, -np.inf], np.nan, inplace=True)
test_df.fillna(0, inplace=True)
```

```python
feature_scale = [feature for feature in train_df.columns 
                   if feature not in ['Id', 'SalePrice'] 
                   and train_df[feature].dtype in ['int64', 'float64']]

# Now apply the scaler only to numeric columns
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_df[feature_scale])

# Note: If you need to process categorical columns separately,
# you would use techniques like one-hot encoding with:
# from sklearn.preprocessing import OneHotEncoder
```

```python
scaler.transform(test_df[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([test_df[['Id']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(test_df[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.to_csv('X_test.csv',index=False)
```

```python
data.to_csv('X_train.csv',index=False)
```
